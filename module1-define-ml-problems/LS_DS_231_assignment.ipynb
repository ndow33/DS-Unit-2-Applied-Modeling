# Begin to clean and explore your data.

```python
df.head(1)
```

```python
test2 = df.loc[0, 'GAME_CLOCK']
test2 = test2.split(':')
test2[0], test2[1]
```

```python
df['MINUTES'] = 0
df['SECONDS'] = 0
df.loc[0, 'MINUTES']
```

```python
counter = 0 

for x in df['GAME_CLOCK']:
  time = x.split(':')
  minute = time[0]
  second = time[1]
  df.loc[counter, 'MINUTES'] = minute
  df.loc[counter, 'SECONDS'] = second
  test = df.loc[counter, 'MINUTES']
  test2 = df.loc[counter, 'SECONDS']
  counter = counter + 1
  print(counter, test, test2)
```

```python
df.loc[0, 'SECONDS'] = int(df.loc[0, 'SECONDS'])
df.loc[0, 'SECONDS']
```

```python
#change seconds and minutes to integers
counter = 0 
for x in df['SECONDS']:
  df.loc[counter, 'SECONDS'] = int(x)
  counter = counter + 1
  print(counter, int(x))

```

```python
df.head()
```

```python
counter = 0 
for x in df['MINUTES']:
  df.loc[counter, 'MINUTES'] = int(x)
  counter = counter + 1
  print(counter, int(x))
```

```python
#remember to account for overtime periods
counter = 0 
df['PERIOD_SECS'] = 0
df['OVERTIMES'] = 0
for x in df['PERIOD']:
  #change x into an integer
  y = int(x)
  #set the value to an integer in the df
  df.loc[counter, 'PERIOD'] = y
  #check to see if the game is in regulation time
  if(y<=4):
    #if so, multiply the remaining quarters by 720 seconds (the time in each 12 minute quarter)
    df.loc[counter, 'PERIOD_SECS'] = 720*(4-y)
  if(y>=5):
    #if the game is in the 5th period or higher, it's in overtime. Set PERIOD_SECS to 0
    df.loc[counter, 'PERIOD_SECS'] = 0
    #find out which overtime the game is in and save as a variable
    overtimes = y-4
    #Save overtimes variable to df
    df.loc[counter, 'OVERTIMES'] = overtimes

  z = df.loc[counter, 'PERIOD_SECS']
  counter = counter + 1
  print(counter, z, overtimes)
```

```python
df['PERIOD'].sort_values(ascending=True)
```

```python
counter = 0 
df['MINUTE_SECS'] = 0
for x in df['MINUTES']:
  y = 60*(x)
  df.loc[counter, 'MINUTE_SECS'] = y
  z = df.loc[counter, 'MINUTE_SECS']
  counter = counter + 1
  print(counter, z)
```

```python
df.head()
```

```python
counter = 0 
df['SECONDS_REMAINING'] = 0
for x in df['SECONDS_REMAINING']:
  a = df.loc[counter, 'SECONDS']
  b = df.loc[counter, 'MINUTE_SECS']
  c = df.loc[counter, 'PERIOD_SECS']
  y = a + b + c
  df.loc[counter, 'SECONDS_REMAINING'] = y
  z = df.loc[counter, 'SECONDS_REMAINING']
  counter = counter + 1
  print(counter, z)
```

```python
df['SECONDS_REMAINING'].sort_values(ascending=True)
```
